```python
import pickle
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
f = open("capacitor_dataset.pickle","rb")
f.seek(0)
data = pickle.load(f)
print(data)
```

```python
data = np.split(data, 2, axis = 1)
print(data)
```

```python
data[1] = np.log(data[1])
print(data)
```

```python
model = LinearRegression()
model.fit(data[0], data[1])
C = np.exp(model.intercept_)/5
R = 1/(-1*model.coef_*C)
print(R,C)
```
